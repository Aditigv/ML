**Q). Assuming that Y is the function of X in the following training set, please try to take second-order hypothesis function to fit (X,Y) coordinate points by curve Y = f(X). Before writing python program to implement regression by gradient descent algorithm, hypothesis function/loss function/cost function are needed for getting all the parameters θs in hypothesis. After training regression module, plot (X,Y) points and the fitting curve by matplotlib python functions**


 q2img.png

```python
import numpy as np
data = np.array([[0,4],[1,5],[2,16],[3,21],[4,36],[5,45],[6,64],[7,77],[8,100],[9,117],[10,144]])
x, Y = data[:,:1], data[:, 1]
from sklearn.preprocessing import PolynomialFeatures
# generate a matrix with all poly combinations that fit the data 
PolyReg = PolynomialFeatures(degree=2)   
X_poly = PolyReg.fit_transform(x) 

pol_reg = LinearRegression()
pol_reg.fit(X_poly,Y)

import matplotlib.pyplot as plt 
plt.scatter(x, Y, color='red')
print("Curve based on given dataset for second degree function=")
print(plt.plot(x, pol_reg.predict(PolyReg.fit_transform(x)), color='blue') )
```

**Q).Write Python program to find the parameters θs in the hypothesis function for the above dataset (dataset in q2) by Cramer’s rule. And compare the results with those coming from gradient descent algorithm**
 


```python
#data
data = np.array([[0,4],[1,5],[2,16],[3,21],[4,36],[5,45],[6,64],[7,77],[8,100],[9,117],[10,144]])
x_, y_= data[:,:1], data[:, 1]
#  matrix x
M = 11   #length of data set
x_bias = np.ones((M,1))
x = np.reshape(x_, (M,1))
x = np.append(x_bias,x,axis=1)
#(xTx)^(-1)
x_transpose = np.transpose(x)   #calculating transpose
product = x_transpose.dot(x)   
#multiplying normal and transposed matrix
temp1 = np.linalg.inv(product) 
print("inverse=\n",temp1) #calculating inverse of product
#xT * y
temp2 = x_transpose.dot(y_)
print("xT * y=""",temp2)
#theta = (xTx)^(-1) * xT * y
theta = temp1.dot(temp2)
print("(xT * y)*inv=""",theta)
print("Now comparing and plot.........")
print("theta0 = """, theta[0])
print("theta1 ="" ", theta[1])
# y = t0 + t1x
plt.scatter(x_,y_)
plt.plot(x, theta[0] + theta[1]*x)
```
